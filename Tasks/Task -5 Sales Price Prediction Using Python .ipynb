### Data Modelling

```python
x = df.drop(['Sales'],1)
x.head()
```

```python
y = df['Sales']
y.head()
```

### Splitting the data into Training and Test data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
model_lr.fit(x_train, y_train)
```

```python
# constant (b - bias)
model_lr .intercept_

# coefficients (w - weights)
model_lr .coef_
```

```python
y_pred = model_lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(rmse)
r2_score=r2_score(y_test,y_pred)
print(r2_score*100)
```

```python
print(y_pred)
```

```python
actual_predict= pd.DataFrame({
    'Actual': y_test.values.flatten(), 
    'Predict': y_pred.flatten()})
actual_predict.head(5)
```

```python
# Calculate the R-squared value for the train data
train_score = model_lr .score(x_train, y_train)

# Calculate the R-squared value for the test data
test_score = model_lr .score(x_test, y_test)

# Print the R-squared values for both the train and test data
print("R-squared value for train data:", train_score*100)
print("R-squared value for test data:", test_score*100)
```
