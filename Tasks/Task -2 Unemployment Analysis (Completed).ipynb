## **Variable description:**

- **`Region`**: The specific region or state to which the data entry belongs. 


- **`Date`**: The date of the data entry, in the format "DD-MM-YYYY.


- **`Year`**: The year corresponding to the data entry. 


- **`Frequency`**: Indicates the frequency at which the data is reported. In this case, it is reported on a monthly basis.


- **`Estimated Unemployment Rate (%)`**: The percentage of unemployed individuals in the specified region and time period. 


- **`Estimated Employed`**: The estimated number of employed individuals in the specified region and time period. 


- **`Estimated Labour Participation Rate (%)`**: The percentage of the population that is participating in the labor force, either employed or actively seeking employment. 


- **`Area`**: Indicates the type of area to which the data corresponds. In this case, it is reported for rural as well as urban areas.


- **`Region-wise`**: Specifies the broader region to which the specific region belongs.

## Data Visualization

### Estimated Number of employees according to different region of india

```python
plt.figure(figsize=(10, 6))

df1['Estimated Employed (Millions)'] = df1['Estimated Employed'] / 1_000_000

ax = sns.barplot(data=df1, x="Region", y="Estimated Employed (Millions)", estimator=sum, ci=None, palette="Set1")
sns.despine(left=True, bottom=True)  

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}M',  
                xy=(p.get_x() + p.get_width() / 2, height), 
                xytext=(0, 3),  
                textcoords="offset points",
                ha='center', va='bottom')

plt.title("Total Estimated Employed (in Millions) by Region")
plt.xlabel("Region")
plt.ylabel("Estimated Employed (Millions)")
plt.show()
```

**Insights**

- Based on the above visual, India's northern region boasts the highest workforce with 1476.89 million employees, followed by the southern region at 1222.30 million. 

- The central region has the smallest workforce with 311.23 million employees, indicating significant regional employment disparities.

```python
# Distribution of Estimated Unemployment Rate
plt.figure(figsize=(8, 6))
sns.histplot(data=df1, x="Estimated Unemployed Rate(%)", kde=True)
plt.title("Distribution of Estimated Unemployment Rate")
plt.xlabel("Estimated Unemployment Rate")
plt.ylabel("Frequency")
plt.show()
```

### Average Unemployment Rate by year

```python
average_unemployment_rate = df1.groupby('Year')['Estimated Unemployed Rate(%)'].mean()
plt.figure(figsize=(5, 4))
plt.pie(average_unemployment_rate, labels=average_unemployment_rate.index, autopct='%1.1f%%', startangle=140)
plt.title('Average Unemployment Rate by Year')
plt.show()
```

**Insights**

- The average unemployment rate in 2020 was exceptionally high at 61.6%, indicating a significant economic challenge. 

- In contrast, the rate decreased to 38.4% in 2019, reflecting a comparatively better employment scenario that year.

### CountPlot: Rural Vs Urban

```python
plt.figure(figsize=(15,4.5))
ax=sns.barplot(x=df1.Area.value_counts() ,y = df1.Area.value_counts().index, data=df1, orient ='h',order = df1.Area.value_counts().index[0:])
plt.xlabel('Count')
plt.ylabel('Area')
plt.title('Count of Employees as per Area',fontsize = 15)
for bars in ax.containers:
    ax.bar_label(bars)
```

### Pie Chart: Regionwise Unemployment Rate Distribution

```python
region_unemployment = df1.groupby("Region")["Estimated Unemployed Rate(%)"].sum().reset_index()
plt.figure(figsize=(5, 4))
plt.pie(region_unemployment["Estimated Unemployed Rate(%)"], labels=region_unemployment["Region"], autopct='%1.1f%%')
plt.title("Unemployment Rate Distribution by Region")
plt.show()
```

**Insights**

- Based on the above visual depicting unemployment rates in India, it's evident that there are substantial regional disparities. 

- The northern region exhibits the highest rate at 40.2%, while the central region stands at the lowest with 2.4%, emphasizing significant variations in unemployment levels across the country.
